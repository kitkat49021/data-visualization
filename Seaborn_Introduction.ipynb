# Seaborn

Seaborn is a library for making statistical graphics in Python. Visualization is an important component of statistical analysis because it can allow us to see trends and patterns that indicate a relationship.

Seaborn is built on top of matplotlib and is closely integrated with pandas data structures. It provides a higher-level of abstraction to make aesthetically appealing statistical visualizations. Seaborn also can be used to simply change the style and appearance of matplotlib visualizations!

# Styles

Styling is the process of customizing the overall look of your visualization, or Figure. Seaborn includes five built-in themes that you can use: `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`.

The default configuration style is `darkgrid`, which you can set by calling the `sns.set_theme()` function without any or parameters (or `sns.set()` which is an alias for `sns.set_theme()`). 

Here, the `darkgrid` style works well since the white grid lines on the grey background don't compete with the lines representing the data.

Seaborn `set_theme` [documentation](https://seaborn.pydata.org/generated/seaborn.set_theme.html)

```python
# Apply the default seaborn style
sns.set_theme()
ax = sns.lineplot(data=tips)
```

To select one of the other the other styles, you can pass its name as a string to the `sns.set_style()` function like we've done here. Here, we're using the `whitegrid` style, which adds a grey grid on a white background. 

Which style should you use? The `whitegrid` and the default `darkgrid` styles both add a grid to your plot, which can be helpful when you want your audience to be able to determine specific values and to draw their own conclusions about the data. Research papers and reports are also a good example of when you would want to include a grid.

Seaborn `set_style` [documentation](https://seaborn.pydata.org/generated/seaborn.set_style.html)

```python
sns.set_style('whitegrid')
ax = sns.lineplot(data=tips)
```

Now, if you are giving a talk or a presentation, it may be better to simplify your graphs by selecting a theme without a grid in order to draw attention to the important visual details in your charts. Seaborn's `dark`, `white` and `ticks` themes don't include a grid.

In the following example, we will use seaborn's `white` theme, which provides clean axes and has a blank background. This style works well when we are more concerned with highlighting the comparisons between groups or the general trend across the groups, and not the specific values.

## simplify the graph using the tick themes not using the grid

```python
sns.set_style('white')
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')
```

The `ticks` theme is similar to `white`; however, it gives a little more structure to the plot by adding small tick marks to the x- and y-axis. 

```python
sns.set_style('ticks')
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')
```

# Context

Seaborn "contexts" don't affect the style of your plots. Instead, these pre-built packages control the scale or size of your plots lines, labels, and other plot elements. The four available options, from smallest to largest, are `paper`, `notebook`, `talk`, and `poster`. The `notebook` style is the default.

If you are giving a presentation and you audience will be farther away, you would want to choose a larger context like `poster`.

Seaborn `set_context` [documentation](https://seaborn.pydata.org/generated/seaborn.set_context.html?highlight=set_context)

```python
sns.set_context('paper')
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.set_context('poster')
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')
```

# Spines

Another way to simplify your plots is by removing the top and right spines from your chart. Each plot has four spines that define data area boundary and connect the axis tick marks.

You can automatically take away the top and right spines using the `sns.despine()` function. However, please note that this function must be called after you have called your plot function!

Seaborn `despine` [documentation](https://seaborn.pydata.org/generated/seaborn.despine.html?highlight=despine#seaborn.despine)

```python
# Reset seaborn style and context
sns.set_theme()

ax = sns.scatterplot(data=tips, x='total_bill', y='tip')

# Remove the top and right spines
sns.despine();
```

You can also control which spines are removed by explicitly setting one or more of the `sns.despine()` function's optional parameters: `left`, `bottom`, `top`, and `right`

Since by default the function removes the `top` and `right` spines, if you want either of them included you will need to set their parameters to `False`. To remove either the `left` or `bottom` spines, you will need to set them to `True`.

```python
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')

# Remove the top and left spines
sns.despine(left=True, right=False)
```

## Sequential color palettes

When plotting numeric data ranging from low to high values, we want to use a sequential color palette, where the primary dimension of variation is luminance. These can be helpful since the equal steps in data are perceived as equal steps in the color space. 

When plotting numeric values, we may want to use a sequential palette that is perceptually uniform. In a perceptually uniform color palettes, the equal steps in data are perceived as equal steps in the color space. Seaborn four perceptually uniform sequential colormaps are `"crest"`, `"flare"`, `"mako"`, and `"rocket"`.  

```python
sns.palplot(sns.color_palette('crest'))
sns.palplot(sns.color_palette('flare'))
sns.palplot(sns.color_palette('mako'))
sns.palplot(sns.color_palette('rocket'))
```

Just like with matplotlib, we can reverse the continuous colormaps by adding the suffix `"_r"` to its name.

```python
sns.palplot(sns.color_palette('flare'))
sns.palplot(sns.color_palette('flare_r'))
```

# Seaborn Functions Overview

As you start to explore seaborn's excellent documentation (which I highly recommend!), you will notice that the plotting methods are structured hierarchically within modules, such as "categorical" and "relational". The plotting methods within each module can be used to achieve similar visualization goals, such as visualizing the distribution of your data.

In addition, each module contains two types of functions. The majority are axes-level functions that will plot your data on to a single `Axes` object, which is the function's return value. In this notebook, I will save the output of these functions as `ax`.

The other type of function is a figure-level function. In contrast to the axes-level functions, these higher-level functions interface with matplotlib most often through a seaborn `FacetGrid`. As you can see in the diagram below, each seaborn module has a single figure-level function (`catplot`, `displot`, `lmplot`, `relplot`) that provides a unitary interface to the axes-level functions within its module.
<img src="../images/seaborn_functions_overview.png" alt="Seaborn Functions Overview" style="width:442px;"/>

The figure-level functions return a `FacetGrid` object, so I will save their return value as `g` to help differentiate from the axes-level plotting methods.

Here is a seaborn's figure-level `displot()` function from the distributions module, which by default will plot a histogram.

Seaborn `displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
# Use displot to create a histogram
g = sns.displot(data=tips, x='total_bill', kind='hist')

# Inspect the data type
type(g)
```

Since the figure-level `displot()` function plots a histogram by default, it is actually using the axes-level `histplot()` function from the distributions module.

The slight different you see between the two plots is because seaborn's figure-level functions plots are square by default. However, they will automatically adjust if needed as we'll see later.

Seaborn `histplot` [documentation](https://seaborn.pydata.org/generated/seaborn.histplot.html)

```python
# Plot a histogram using the axes-level function
ax = sns.histplot(data=tips, x='total_bill')

type(ax)
```

You can easily access any of the other axes-level functions in the module by passing their name to the `kind` parameter.

| Name | Axes-level function | Plot Type |
| :--- | :-----------------: | :-------- |
| `"hist"` | `histplot()` | Histogram (default) |
| `"kde"` | `kdeplot()` | Kernel Density Estimation (KDE) | 
| `"ecdf"` | `ecdfplot()` | Empirical Cumulative Distribution Functions (ECDF) (univariate-only) |

In addition, another difference you'll notice is that figure-level functions will automatically place the legend ouside of the plot.

Seaborn `displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)  
Seaborn `kdeplot` [documentation](https://seaborn.pydata.org/generated/seaborn.kdeplot.html)

```python
# Create a KDE plot
g = sns.displot(data=tips, x='total_bill', hue='sex', multiple='stack', kind='kde')
```

The figure-level functions most useful feature is that they let us easily create Figures with multiple subplots. This is often referred to as "facetting". For example, instead of stacking both histograms in the same plot, we can plot each distribution in its own subplot.

```python
g = sns.displot(data=tips, x='total_bill', col='smoker')
```

Also, I want to point out here that there are different ways we can pass in our data to seaborn's plotting functions. Now, the first example below does work but it is not recommended. Instead, you should follow the convention in the second example.

# Figure Size

If no Axes object is explicitly provided, seaborn's axes-level functions will use the currently active Axes, so the default plots will have the same size as the underlying matplotlib functions. In these cases, if we want to control the Figure size, we need to create it ourselves using setting the optional `figsize` parameter with either `plt.figure()` or `plt.subplots()`.

```python
plt.figure(figsize=(6,3))
ax = sns.histplot(data=tips, x='total_bill');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))

ax1 = sns.histplot(data=tips, x='total_bill', ax=ax1)
ax2 = sns.histplot(data=tips, x='tip', ax=ax2)
```

For the higher-level, figure-level functions, we need to use the `aspect` and `height` parameters to change the size of our Figure, since these functions are often used to create multiple subplots. It's important to remember that these parameters apply to each subplot separately and not to the Figure as a whole.

To help see how they work, here I've used `lmplot()` to create two facets (each subplot is a facet). Since the `"sex"` column has two unique values, it automatically created a `FacetGrid` with 2 subplots. If instead we pass the `"day"` column to the `col` parameter, the function would create 4 facets since it has four unique values.

Seaborn `FacetGrid` [documentation](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html)  
Seaborn `lmplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

```python
# Plot a FacetGrid using the defaults
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='sex')
# g = sns.lmplot(data=tips, x='total_bill', y='tip', col='day')
```

`height` represents the height of facets (subplots) in inches and `aspect` is the ratio of width and height (`width = aspect * height`). By default, `lmplot()` has a `height` is set to `5` and `aspect` is set to `1`.

Now let's see what happens when we reduce the `height` to `3`.

```python
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', height=3)
```

The `aspect` is set to `1` by default. If we set it to `0.5`, our facets will now be twice as tall as they are wide.

```python
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', aspect=0.5)
```

However, if we set `aspect` to `2`, now our plots are twice as wide as they are tall.
## Make graphs wide

```python
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', aspect=2)
```

# Categorical Plots

In seaborn, there are several different ways to visualize a relationship involving categorical data.

Seaborn figure-level `catplot()` function gives us access to eight axes-level plotting methods that show the relationship between numerical variable and one or more categorical variables. We can access them via `catplot()` by specifying it with the `kind` parameter.

Catplot can handle 8 different plots currently available in Seaborn. catplot function can do all these types of plots and one can specify the type of plot one needs with the kind parameter. If we don't specify the type of plotting method to use, `catplot()` will create a strip plot by default by calling `sns.stripplot()`.

## Count Plots

For the categorical plotting methods, let's start by looking at `countplot()`, which is an axes-level function that visualizes the number of occurences for each unique category value.

Notice how since we are using a named column from our DataFrame, seaborn automatically adds the column name as the x-axis label at the bottom.

Seaborn `countplot` [documentation](https://seaborn.pydata.org/generated/seaborn.countplot.html)

```python
# Show count of observations
ax = sns.countplot(x='class', data=titanic)
```

## Bar Plots

Seaborn bar plots are used to plot the mean of a numerical column against all of the unique values in a categorical column. You should pass the name of your categorical variable to `x` and the numerical variable to `y`. 

By default, `barplot()` uses `"mean"` to estimate the central tendency of each category against the numeric column we've passed to `y`. For example, here we can see that the average survival rate for female passengers was more than 70%, while for men the average survival rate was less than 20%. 

We could calculate the height of each bar by grouping the original DataFrame columns and then averaging the `"sex"` column.

Seaborn `barplot` [documentation](https://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
titanic[['sex', 'survived']].groupby('sex').mean()
```

```python
ax = sns.barplot(x='sex', y='survived', data=titanic)
```

We can use the method's optional `hue` attribute to group the bar plots even further into sub-categories. For example, here we've plotted the passengers survival rate with respect to gender, which is then further divided into class.

Now we can see that the survival rate was highest for the females with first class tickets, followed by second and third class. In addition, we can also see the same trend for the male passengers. 

```python
ax = sns.barplot(data=titanic, x='sex', y='survived', hue='class')
```

We can use a different color palette by passing its name to the optional `palette` parameter.

```python
ax = sns.barplot(data=titanic, x='sex', y='survived', hue='class', palette='Set2')
```

Or, if you want to specifically set each other, you can create a dictionary that maps each subgroup to a color and then pass it to the `palette` parameter.

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {'First': '#ff7f0e', 'Second': '#d62728', 'Third': '#2ca02c'}

ax = sns.barplot(data=titanic, x='sex', y='survived', hue='class', palette=palette_colors)
```

By default, seaborn's `barplot()` calculates the mean values. However, we can display the sum of the numerical values per category by setting `estimator=sum`. Some other available options you might find useful are passing NumPy's `np.median` or `np.std` functions to the `estimator` parameter in order to find the median or standard deviation values respectively.

To update the default x- and y-axis labels, one quick way is to chain matplotlib's `set()` method to the `barplot()` function.

Now, our plot is showing, for both genders, the sum of surviving passengers and we can see that more than 80 female first-class passengers survived. The lowest number of surviving passengers were males traveling in second class.

Matplotlib `Artist.set` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.artist.Artist.set.html)

```python
ax = sns.barplot(data=titanic, x='sex', y='survived', hue='class', estimator=sum
                 ).set(xlabel='Gender', ylabel='Survival Rate')
```

If we wanted to switch the order of our columns, we could do this by passing a list to the `order` parameter.

```python
ax = sns.barplot(data=titanic, x='sex', y='survived', hue='class', order=['female', 'male'])
```

By default, `barplot` will also compute a bootstrap 95% confidence interval for each calculated value, which are plotted as black, vertical lines.

In the plot above, the confidence intervals show that if the dataset contained a random sample of Titanic passengers, the difference between passengers' age for those that survived and who did not is not statistically significant at the 5% significance level.

If this information is not needed, you can hide these bars by setting `ci=False`. You may also want to do this when working with larger datasets, since the confidence levels can take additional time to compute!

```python
ax = sns.barplot(data=titanic, x='sex', y='age', ci=False)
```

You can also show the standard deviation of observations instead of a confidence interval. Here, I've also added caps to the error bars using the `capsize` parameter.

```python
ax = sns.barplot(data=titanic, x='sex', y='age', ci='sd', capsize=0.2);
```

Seaborn's `catplot` is a Figure-level interface for drawing categorical plots onto a `FacetGrid`. By default, it will plot a strip plot. However, we can use it create a bar plot by passing `"bar"` to the `kind` parameter, which selects which underlying axes-level function to use. This allows us to plot multiple subplots that correspond to the unique values from the categorical column (`"alone"`) that we've passed to the optional `col` parameter.

For example, here we've plotted gender against the survival rate for passengers traveling alone. On the left, we can see the gender vs survival rate for passengers not traveling alone, and on the right we can see the same information for passengers who were traveling alone.

Seaborn `catplot` [documentation](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
g = sns.catplot(
        data=titanic, x='sex', y='survived', height=3, ci=False, kind='bar', hue='class', col='alone')
```

Since the `"class"` column contains 3 unique values, `catplot()` will automatically create a row with 3 subplots when we pass the column name to the `col` parameter. However, we can also tell seaborn to only have 2 columns per row by setting `col_wrap=2`. 

```python
g = sns.catplot(data=titanic, x='sex', y='survived', height=3, col='class', kind='bar', col_wrap=2)
```

## Point Plots

Similar to bar charts point plots can also be useful to visualize how values change across categorical intervals. Instead of plotting bars, the points mark a single point at the end of where a bar would go.

They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same `hue` level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.

Also similar to the `barplot()` method, `pointplot()` will automatically group the data and then calculate the average value, with the 95% confidence interval plotted as vertical lines centered on each point.

Seaborn `pointplot` [documentation](https://seaborn.pydata.org/generated/seaborn.pointplot.html)

```python
# Instantiate the Figure and a 1x2 grid
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

# Plot a bar plot
sns.barplot(data=titanic, x='sex', y='survived', hue='class', ax=ax1
                 ).set(xlabel='Gender', ylabel='Survival Rate')

# Plot a point plot
sns.pointplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2
                 ).set(xlabel='Gender', ylabel='Survival Rate');
```

## Histograms

Seaborn offers functions to examine univariate and bivariate distributions. The `displot()` method is a figure-level function that is drawn onto a `FacetGrid` and it plots a histogram by default. Here, we will use it to look at a univariate distribution.

It is relatively new to seaborn and should not be confused with the now deprecated `distplot()` method.

Seaborn `displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
g = sns.displot(tips, x='total_bill')
```

We can also use histograms to visualize the distribution of the categorical variables in our dataset. Conveniently, seaborn will automatically select discrete bins when plotting categorical variables.

However, you may find it helpful to slightly reduce the size of the bars since the x-axis is not a continuous number like in the previous example. This helps to emphasize that the values along your x-axis are categorical. We can do this by setting the `shrink` parameter to a value less than `1`, which is the default.

```python
# g = sns.displot(tips, x='day')
g = sns.displot(tips, x='day', shrink=0.9)
```

When your data only contains a relatively small number of integer values, you'll often find that the default bin width will be too small. 

```python
# Use default bin width
g = sns.displot(data=tips, x='size')
```

In this case, we can either pass array to the `bins` parameter specifying the precise bin breaks, or we can set `discrete=True` so that the plot will use the unique values in the data to choose the bin breaks and will center the bars over their corresponding value. 

```python
g = sns.displot(data=tips, x='size', discrete=True)
# sns.displot(data=tips, x='size', bins=[1, 2, 3, 4, 5, 6, 7])
```

Often we will want to see if a variable's distribution differs across other variables in our dataset. Here, we can do this by using the optional `hue` parameter. Now, seaborn plots a separate histogram for each unique value.

```python
g = sns.displot(data=tips, x='total_bill', hue='sex')
```

### Step Plot

By default, seaborn will layer the histograms on top of each other which can make it difficult to distinguish them. When this happens, option is to change the visual representation of the histogram to a step plot.

```python
g = sns.displot(data=tips, x='total_bill', hue='sex', element='step')
```

### Stacked Histogram

Or an alternative approach is to create a stacked histogram by setting `multiple="stack"`.

```python
g = sns.displot(data=tips, x='total_bill', hue='sex', multiple='stack')
```

To ensure that there are no overlaps and that the bars remain comparable in terms of height, we can set `multiple="dodge"`.

However, keep in mind that this only works well when the categorical variable has a small number of levels or unique values, like we have here.

```python
g = sns.displot(data=tips, x='total_bill', hue='sex', multiple='dodge')
```

Since `displot()` is figure-level function, we can draw each individual distribution in a separate subplot by passing in a second column to either the `col`or the `row` parameter.

```python
g = sns.displot(data=tips, x='total_bill', col='sex')
```

# Normalize

It can be difficult to compare distributions when the subsets in our data have an unequal numbers of observations. In this case, we can normalize the counts setting the optional `stat` parameter to `"density"`, and set `common_norm=False` so that each subset is normalized independently of each other.

```python
g = sns.displot(tips, x='total_bill', col='sex', stat='density', common_norm=False)
```

Also new to seaborn is `histplot`, an axes-level function that is recommended for when you need to plot a histogram onto an existing matplotlib Figure.

Here, we will use this new function to look at how setting the width of each bin with the optional `binwidth` parameter can affect our plot. This parameter is automatically set by seaborn; however, it can often be helpful to experiment with different sizes.

Seaborn `histplot` [documentation](https://seaborn.pydata.org/generated/seaborn.histplot.html)

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Show the histogram with the default values
sns.histplot(tips, x='total_bill', ax=ax1)

# Reduce the width of the bins
sns.histplot(tips, x='total_bill', binwidth=2, ax=ax2);
```

We can also explicitly set the number of bins to use by setting the `bins` parameter.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

sns.histplot(tips, x='total_bill', ax=ax1)
sns.histplot(tips, x='total_bill', bins=20, ax=ax2);
```

# Relational Plots

Seaborn has two axes-level plotting methods to visualize the relationships between your variables, `lineplot()` and `scatterplot`, along with the higher-level `relplot()` that can be used to access them both.

## Line Plot

Line plots visualize the relationship between two columns in the form of a line. 

Here, we'll plot a line plot that shows the relationship between the `"size"` and `"total_bill"` columns by passing in the name of our DataFrame to the `data` argument and the column names to `x` and `y`.

In our plot, we can see that as the number of people in the party increased, so did the total amount of the bill.

Seaborn `lineplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill')
```

### Confidence Interval

In our previous plot, the semi-transparent region surrounding the line is the confidence interval. However, it can be removed by setting the `ci` parameter of the `lineplot()` to `None` as we've done here.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill', ci=None)
```

By default, the confidence area is represented by a translucent area around the line. However, we can choose to have it represented in the form of vertical bars by passing `"bars"` to the `err_style` argument.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill', err_style='bars')
```

**Hue Parameter**

One big advantage of using seaborn is that we can easily add a third variable to our plot using the `hue` parameter. The `hue` parameter allows us to express a third dimension of information using color. 

We can also plot a line for each unique value in a categorical column. We do this by specifying the column value for the `hue` parameter. For example, here our plot includes a line showing the relationship between size of the party and the total bill for smokers, while the other line plot displays the same information for non-smokers.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill', hue='smoker')
```

### Line Plot Style

To help differentiate between multiple lines in a line plot, we can use the optional `style` parameter and set its value to the name of the column value we passed to `hue`. Now, we can see that the line plot for non-smokers is now dashed.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill', hue='smoker', style='smoker')
```

### Line Markers

We can also add markers to our line plot. Remember that markers are special symbols that appear at the places in a line plot where the values for x and y axes intersect. To plot markers, you have to pass a list of symbols in a list to the markers attribute. Each symbol corresponds to one line plot.

Here, we can create a line plot that uses circles and triangles as markers.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill', hue='smoker', style='smoker', markers=['o', '<'])
```

### Line Plot Colors

In addition to changing the line style, we can change the colors in our plot by passing in the name of a color palette to the optional `palette` parameter. Here, let's try using the `"hot"` palette.

```python
ax = sns.lineplot(data=tips, x='size', y='total_bill',  hue='smoker', palette='hot')
```

# Regression Plots
 
Two main functions in seaborn are used to visualize a linear relationship between our variables, `regplot()` and the figure-level `lmplot()`.

By default, both functions draw a scatter plot of the two variables, `x` and `y`, and then fit the linear regression model and plot the resulting regression line with a 95% confidence interval.

Seaborn `regplot` [documentation](https://seaborn.pydata.org/generated/seaborn.regplot.html)  
Seaborn `lmplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

```python
ax = sns.regplot(data=tips, x='total_bill', y='tip')
```

**Faceting multiple regressions**

While `regplot()` allows us to explore the relationship between a pair of variables, it is only able to show a single relationship. However, since `lmplot()` combines `regplot()` with `FacetGrid`, it provides us a convenient interface to explore whether the relationship between these two variables changes as a function of a third variable. 

Like seaborn's other figure-level functions, we can organize data by color using the `hue` parameter and by an additional column using the `col` or `row` parameter.

Seaborn `lmplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

```python
g = sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

We can create a subplot for the lunch and the dinner time crowds by setting `col="time"`.

```python
g = sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', col='time')
```

## Residuals Plot

We can use seaborn's `residplot()` function to evaluate whether our regression model is appropriate for our dataset. This axes-level function fits and removes a simple linear regression and then plots the residual values for each observation. Ideally, these values should be randomly dispersed around `y=0`.

Seaborn `residplot` [documentation](https://seaborn.pydata.org/generated/seaborn.residplot.html)

```python
ax = sns.residplot(data=tips, x='total_bill', y='tip')
```

## Heatmaps

A heatmap is a matrix-like plot that we can use to visualize the degree of correlation between multiple numerical columns, displays higher values with more intense color.

In our first example, we will call `corr()` on our dataset, which returns the correlation between all of its numeric columns. We can also plot the correlation values on our heatmap by passing `True` to the `annot` parameter.

Seaborn `heatmap` [documentation](https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap)

```python
plt.figure(figsize=(8,8))
ax = sns.heatmap(titanic.corr(), cmap='coolwarm');
# ax = sns.heatmap(titanic.corr(), cmap='coolwarm', center=0, vmin=-1, vmax=1, annot=True);
```

```python
plt.figure(figsize=(9,6))
ax = sns.heatmap(data);
# ax = sns.heatmap(data, linewidth=1.1, linecolor='darkgrey');
```
